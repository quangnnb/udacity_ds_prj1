# Exploring Salary Trends in India

In the fast-paced world of job markets, understanding salary trends is key to making informed career choices. In this step-by-step guide, we'll uncover the answers to three intriguing questions about salaries in India in 2022:

1. Which are the top 10 companies that pay the highest salaries?
2. Which are the top 5 cities/regions with the highest average salaries?
3. Which are the top 5 job titles with the highest salaries?
Let's embark on this data-driven journey together!

Prepare Data
To get started, let's import our libraries.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly
```

For this project I decided to analyse the Salary dataset found on Kaggle: https://www.kaggle.com/datasets/iamsouravbanerjee/software-professional-salaries-2022

Let check the head of the dataset.

```python
df = pd.read_csv("Salary_Dataset_with_Extra_Features.csv")
df.head()
```

```python
#display the columns of data frame
df.columns
```

```python
# check null if any in the df
df.isnull().sum()
```

```python
# make sure that there is no duplicate in the dataset
df.duplicated().sum()
```

First, I would like to discover which job title has the highest salary.

```python
# showing that the relative between job roles and salary
role_vs_sal = pd.DataFrame(df.groupby("Job Roles")["Salary"].mean().round(0)).reset_index()
role_vs_sal.sort_values("Salary", axis = 0, ascending = False, inplace = True)
```

To describe the relationship between job titles and salaries, I will create a bar chart to provide a clearer visual representation.

```python
# Create a bar chart using matplotlib.pyplot
plt.figure(figsize=(12, 6))
plt.bar(role_vs_sal["Job Roles"], role_vs_sal["Salary"], color='green')
plt.xlabel("Job Roles")
plt.ylabel("Salary")
plt.title("Job roles and average salary")
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()

plt.show()
```

It shows that Database, Mobile, and SDE are the top three highest-paid roles in India in 2022.

Now continuing to figure out our first question.

## 1. Which are the top 10 companies that pay the highest salaries?


```python
# using pandas to create a dataframe contain 10 companies that pay the highest salaries.
top_10_companies = pd.DataFrame(df[['Company Name', 'Salary']].nlargest(10, 'Salary'))


# showing the data via the bar chart
plt.figure(figsize=(12, 6))
plt.bar(top_10_companies["Company Name"], top_10_companies["Salary"], color='skyblue')
plt.xlabel("Company Name")
plt.ylabel("Salary")
plt.title("Top 10 Companies with the Highest Salaries")
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()

plt.show()
```

Consequence: It's showing that Tharpar Universiry is the highest-paid salary in the dataset, following by Concentrix, Koru UX Design,...

## 2. Which are the top 5 cities/regions with the highest average salaries?

With the second question, we will find out the average of cities or regions and what is the top highest salary there.

```python
# 2. Which are the top 5 cities/regions with the highest average salaries?
top_5_cities = df.groupby('Location')['Salary'].mean().round(0).nlargest(5).reset_index()
top_5_cities.rename(columns={'Salary': 'Average Salary'}, inplace=True)

print("\nTop 5 cities/regions with the highest average salaries:")
print(top_5_cities)

# Create a bar chart using matplotlib.pyplot
plt.figure(figsize=(12, 6))
plt.bar(top_5_cities["Location"], top_5_cities["Average Salary"], color='gray')
plt.xlabel("Location")
plt.ylabel("Salary")
plt.title("Top 5 Cities/Regions with the Highest Average Salaries")
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()

plt.show()
```

Mumbai is one of the major cities in India, so it's easy to understand that it has the highest average salary.

## 3.Which are the top 5 job titles with the highest salaries

And finally, which jobs title are the most attractive job in India in 2022, Let check it out!

```python
# 3.Which are the top 5 job titles with the highest salaries
top_5_jobs = df[['Job Title', 'Salary']].nlargest(5, 'Salary')
print("Top 5 job titles with the highest salaries:")
print(top_5_jobs)

# Create a bar chart using Plotly
plt.figure(figsize=(12, 6))
plt.bar(top_5_jobs["Job Title"], top_5_jobs["Salary"], color='red')
plt.xlabel("Location")
plt.ylabel("Salary")
plt.title("Top 5 Job Titles with the Highest Salaries")
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()

plt.show()
```

Undoubtedly, SDE (Software Development Engineer) is the hottest job in India in 2022. India's IT market is continuously expanding, and there is no sign that its development will stop growing.
